# A* SEARCH ALGORITHM - DESCRIPTION

* Reference : https://en.wikipedia.org/wiki/A*_search_algorithm <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; This computer algorithm is an algorithm that is widely used in pathfinding and graph traversal. In essence, it is Dijkstra's algorithm with a heuristic. Disjkstra's algorithm traverses a graph from the starting node and keeps a list of its neighbors and the distances to them. Then it chooses the node with the smallest distance in this list and records the neighbors of this node while also recording the total distance to these neighbors from the starting point. This is repeated until the node chosen from the list is the destination. Because the algorithm chooses the next node to visit according to the smallest total distance to that node, it ensures that all possible paths are considered to guarantee optimality due to multiple instances of a node being allowed on the list with different total distances. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The A* search algorithm is a best first search algorithm which chooses the most promising path based on a heuristic. This heuristic function estimates the Manhattan or Euclidean distance left to the destination and keeps a score for every node it visits. Then based on these scores, the algorithm extends one node at each iteration, keeping a tree of paths until reaching its goal. The idea is the same as Dijkstra's but in this case, the heuristic allows the algorithm to choose the next node in a more promising fashion rather than blindly picking the node with the smallest total distance from the list. Here our f_score acts as the distance to be recorded in Disjkstra's, the g_score is the weight of current node to its neighbor, and the h_score is the value based on our heuristic. Then our algorithm chooses the next promising node from the list according to the lowest f_score which equals the sum of the g_score and the h_score. Therefore, it allows for a more direct search as long as the heuristic is monotone, h(x) $\le$ d(x,y) + h(y), which means that no node has to ever be visited twice.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Some particulars about our algorithm are the use of a node class, a priority queue, sets, and an adjacency list of neighbors.

* Node Class
    * Allows for storage of multiple variables within a node.
    * This allows for faster access of necessary data and better organization.
    
* Priority Queue
    * Used as the data structure for the above mentiond "list" in Dijkstras's.
    * A priority queue keeps track of the smallest distance value in the list for fast access - O(1).
    * Time complexity of the algorithm is greatly reduced due to this function from O(log(n)).
    
* Sets
    * A set is used to update existing nodes and keep track of visited nodes in A*.
    * This data structure is good at finding existing nodes within its list by nature and greatly reduces time in its search.
    * It also has a similar function to the priority queue of keeping track of the priority value.
    
* Adjacency List
    * An adjacency list of neighbors keeps memory down to a minimum size of O(V*E).
    * An adjacency matrix would be O(V*V).

```python
import math
from heapq import heappush, heappop
# A* Algorithm

# 1. Neighbors_and_weights can be an adjacency / weights list
#   a. Weights will hold temporal value
#   b. Want to reduce memory by implementing list of arrays rahter than an nxn matrix.
#      weights[node[i].key] = ((neighbor1.key, weight), (neighbor2.key, weight), ... )
# 2. An array pathTO will be used to keep track of our optimal route

v = 10000                              # length of nodes - test value 5
neighbors_and_weights = v*[]        # TBD
pathTo = v*[-1]                     # keeps track of optimal path

class Node:                                                  # the node's index in the list will be its key
    def __init__(self, key, x_coord, y_coord, neighbors_and_weights = []):  
        self.key = key                                       # needed to record pathTo
        self.x_coord = x_coord                               # coordinates to be used to determine distance(probably)
        self.y_coord = y_coord
        self.neighbors_and_weights = neighbors_and_weights   # neighbors and weights adjacency list
        self.g_score = 'inf'                                 # score of path from starting point to this vertex - TBD      
        self.f_score = 'inf'                                 # f_score = g_score + h_score
        self.landmark_distances = []                         # distances to all landmarks (landmark, distance)
        self.landmark_closest = -1                           # index to closest landmark - default 0
        
        # a. Maybe we could add another property here for pre-processing(something like "reach")
        #    Possibly employ some kind of density value of time per unit length since we have a time based metric
        # b. Can also add other values here as we need
            
def Heuristic(current_node, terminal_node):  # determines h_score - score of path from this vertex to goal - Euclidean
    current = current_node    # initial walking speed ~ 30 seconds per 50 meters(unit length) 
    goal = terminal_node      # can change this according to our map and increasing number of nodes

    Euclidean = math.sqrt(abs(current.x_coord - goal.x_coord)**2 + abs(current.y_coord - goal.y_coord)**2)
    h_score = Euclidean * 0.5            # half a minute(30 seconds) per unit length
    
    return h_score                       # so all our scores will be in minutes


def retrieve_path(pathTo, start, end):     # returns path
    parent = pathTo[end]
    path = [end]
    print end
    while  parent != start:
        print parent
        path.append(parent)
        parent = pathTo[parent]
    print start
    path.append(start)          # appending the starting point
    
    # return the full path from the end to start
    return path
    
# 3. F_score = g_score + h_score will be determined at each iteration
# 4. Will use "sets" as our data structure because of its innate prowess at finding if a key exists in the set
#   a. Sets are also good for finding the min value, which we will use for f_score
# 5. Will need to check for out of bounds (starting and ending points)

def A_star(Nodes, start, end):
    if start < 0 | start > v:
        return "The starting point doesn't exist in UCLA!"
    if end < 0 | end > v:
        return "The end point doesn't exist in UCLA!"
    
    closed_set = set()     # will store nodes that are visited
    open_set = set()       # will store nodes that are neighbors in the optimal path (minimaum f_score)
    
    goal = Nodes[end]
    entry = Nodes[start]                     # starting node
    entry.g_score = 0                        # starting node g_score is 0
    entry.f_score = Heuristic(entry, goal)   # starting node f_score = g_score + h_score 
    open_set.add((entry.f_score, entry))
    
    while (len(open_set) > 0):
        current = min(open_set)[1]       # pick node with lowest f_score
        
        if current == goal:
            return current.f_score      # returns estimated time - **can a function return 2 different value types?**
        
        open_set.remove((current.f_score, current))        # remove chosen node from open_set
        closed_set.add(current)                           # add chosen node to closed_set
        
        for i in current.neighbors_and_weights:
            if Nodes[i[0]] in closed_set:    # if Nodes[neighbor.key] is in closed set - do nothing
                continue                     
                                             # i[1] = weight of edge between current and neighbor
            g_new = current.g_score + i[1]   # compute new g_score
            
            if (Nodes[i[0]].f_score,Nodes[i[0]]) in open_set:  # if there is a duplicate of Nodes[neighbor.key] in open_set
                if Nodes[i[0]].g_score > g_new:   # compare new g_score and update if new score is lower
                    Nodes[i[0]].g_score = g_new
                    
            else:
                Nodes[i[0]].g_score = g_new                                 # if in neither, update g_score
                Nodes[i[0]].f_score = g_new + Heuristic(Nodes[i[0]], goal)  # calculate h_score and f_score
                pathTo[i[0]] = current.key                                     # record path to parent
                open_set.add((Nodes[i[0]].f_score, Nodes[i[0]]))            # add to open_set
                
    # Will not need to return anything here as our graph will be connected and will therfore always find a path.
    
```

## Testing against Dijkstra's

```python
from heapq import heappush, heappop

def dijkstra(nodes,start,end):
    distances = [float('inf') for i in range(len(nodes))] #initialize distances
    parents = [-1 for i in range(len(nodes))]
    distances[start]=0
    que=[(distances[start],start)]
    while que:
        state=heappop(que)[1]
        if state==end:
            break
        for neigh in nodes[state].neighbors_and_weights:
            dist_from_state=distances[state]+neigh[1] # neigh[1] = weight of neigh
            if dist_from_state<distances[neigh[0]]:
                distances[neigh[0]]=dist_from_state                
                parents[neigh[0]]=state #Keeps track of the parent
                if distances[neigh[0]]<float('inf'):
                    heappush(que,(distances[neigh[0]],neigh[0]))
    if distances[end]==float('inf'):
        print "No path"
        return [-1,[]]
    path=[end]
    cur_par=end
    while cur_par!=start:
        path.append(parents[cur_par])
        cur_par=parents[cur_par]
    return [distances[end],path]
```

## Testing against A*2 - Disjkstra's with a Heuristic

```python
from heapq import heappush, heappop
import math

def h_score(curnode,endnode,effect):
    return effect*math.sqrt((curnode.x_coord-endnode.x_coord)**2+(curnode.y_coord-endnode.y_coord)**2)

def Astar_2(nodes,start,end,effect=0.5):
    parents = [-1 for i in range(len(nodes))]
    g_score = [float('inf') for i in range(len(nodes))]
    f_score = [float('inf') for i in range(len(nodes))]
    g_score[start]=0
    f_score[start]=h_score(nodes[start],nodes[end],effect)
    que=[]
    heappush(que,(f_score[start],nodes[start]))
    while que:
        cur=heappop(que)[1]
        if cur.key==end:
            break
        for neigh in cur.neighbors_and_weights:
            g_score_neigh=g_score[cur.key]+neigh[1] # neigh[1] = weight of neigh
            if g_score_neigh<g_score[neigh[0]]:
                g_score[neigh[0]]=g_score_neigh
                f_score[neigh[0]]=g_score_neigh+h_score(nodes[neigh[0]],nodes[end],effect)
                parents[neigh[0]]=cur.key #Keeps track of the parent
                if f_score[neigh[0]]<float('inf'):
                    heappush(que,(f_score[neigh[0]],nodes[neigh[0]]))
    path=[end]
    cur_par=end
    while cur_par!=start:
        if parents[cur_par]==-1:
            return 'No path'
        path.append(parents[cur_par])
        cur_par=parents[cur_par]
    return [g_score[end],path]
```

## Testing on Square Graph with Weighted Diagonals + Varying Heuristic

```python
# side weight < diagonal weight
def seqgraph2(n,side_effect,diagonal_effect): # create n**2 nodes graph
    nodes=[]
    index=0
    side=1*side_effect # weight for side edges
    diagonal=1.4142*diagonal_effect # weight for diagonal edges, sqrt(2)=1.4142
    for x in range(n):
        if x!=0 and x!=n-1:
            node=Node(index,x,0,[])
            node.neighbors_and_weights=[(index-1,side),(index+1,side),(index+n-1,diagonal),(index+n,side),(index+n+1,diagonal)]
            nodes.append(node)
        elif x==0:
            node=Node(index,x,0,[])
            node.neighbors_and_weights=[(index+1,side),(index+n,side),(index+n+1,diagonal)]
            nodes.append(node)
        elif x==n-1:
            node=Node(index,x,0,[])
            node.neighbors_and_weights=[(index-1,side),(index+n,side),(index+n-1,diagonal)]
            nodes.append(node)  
        index=index+1
    for y in range(1,n-1):
        for x in range(n):
            if x!=0 and x!=n-1:
                node=Node(index,x,y,[])
                node.neighbors_and_weights=[(index-n-1,diagonal),(index-n,side),(index-n+1,diagonal),(index-1,side),(index+1,side),
                                            (index+n-1,diagonal),(index+n,side),(index+n+1,diagonal)]
                nodes.append(node)
            elif x==0:
                node=Node(index,x,y,[])
                node.neighbors_and_weights=[(index-n,side),(index-n+1,diagonal),(index+1,side),(index+n,side),(index+n+1,diagonal)]
                nodes.append(node)
            elif x==n-1:
                node=Node(index,x,y,[])
                node.neighbors_and_weights=[(index-n-1,diagonal),(index-n,side),(index-1,side),(index+n-1,diagonal),(index+n,side)]
                nodes.append(node)
            index=index+1
    for x in range(n):
        if x!=0 and x!=n-1:
            node=Node(index,x,n-1,[])
            node.neighbors_and_weights=[(index-n-1,diagonal),(index-n,side),(index-n+1,diagonal),(index-1,side),(index+1,side)]
            nodes.append(node)
        elif x==0:
            node=Node(index,x,n-1,[])
            node.neighbors_and_weights=[(index-n,side),(index-n+1,diagonal),(index+1,side)]
            nodes.append(node)
        elif x==n-1:
            node=Node(index,x,n-1,[])
            node.neighbors_and_weights=[(index-n-1,diagonal),(index-n,side),(index-1,side)]
            nodes.append(node)  
        index=index+1
    return nodes
```

# ANALYSIS OF TIME COMPLEXITY

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The time complexity of the algorithm depends heavily on its heuristic function because the algorithm chooses its path partly based on the lowest heuristic score in the set. And because it is an algorithm that uses a heuristic with Dijkstra's algorithm, it has a lower time complexity than that of Dijkstra's. While the time complexity of Dijkstra's is O(nlog(n)), the time complexity of A* can be seen as O(h) where h is the number of times our heuristic is called. Disjkstra's uses a heap to append the neighbors of every node it visits and chooses the minimum cost within that heap, but in addition to this, A* employs a heursitc that allows for a more informative subsequent node selection process. Whereas Dijkstra's could find a very roundabout path to our goal, A* could find a more direct path without compromising optimality.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; These time complexities are all based on the fact that we used the most optimal data structure in each function. Without these we would not be able to reach the time complexities mentioned above. For example, if we did not use a priority queue to keep track of the minimum total distance to each node from the starting point, the algorithm would have to find the node with the smallest total distance by traversing all the nodes in the list at each iteration, increasing the time complexity from O(1) to O(log(n)). This would make the algorithm unusable. Also using an ajacency list to keep memory size down to a minimum allows for a less unruly data structure.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Another factor in memory reduction worth mentioning is the use of an open set and a closed set in the A* algorithm. In Dijkstra's, a priority queue is used to access the node with the minimum total distance in the list. However, because the algorithm allows for multiple instance of the same node to exist within this list at any given time, the size of this list could grow rapidly in a very densely connected graph. A* can keep the size of its list to a minimum by using a set for this list instead or a priority queue. In addition to having the ability to keep track of the priority node, when a duplicate instance of the same node is introduced, a set finds the existing node within itself and updates it rather than adding another instance. This is possible because of the set's inherent ability to quickly find existing nodes within itself, whereas this is not possible in the same time complexity with a priority queue. This keeps the size of the set down to a minimum, never exceeding O(V) in size. This comes in handy for extremely large and densely connected graphs, which may produce cases where nodes are visited multiple times in its traversal.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; We also conclude that A-2, Disjktra's with a heurisitc, is faster than Dijkstra's and A* in determining the fastest path.

# PRE-PROCESSING

* Reference : https://archive.siam.org/meetings/alenex04/abstacts/rgutman1.pdf

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; We now further our study and hope to modify our algorithm in an attempt to give it the ability to tackle graphs that are more dense as well as dynamic. Also, our herusitic so far is determined by calculating the euclidean distance between the subject pair of nodes. But what if the weight of our edges had a time metric? Then our heurisitic would serve to be inaccurate. We can also address this particular issue with pre-processing.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Historically in industry, two main strategies were used for high volume graphs. The first of such methods avoids an all pairs computation by only computing paths that can connect the starting and final points with a number of pre-computed paths. This type of "pre-processing" partitions the graph into sub-paths but takes too long to process while having to assume planarity and only being able to partition undirected graphs. The other exploits the natural hierarchy of a graph and gives edges priority based on the distance needed to travel. ## Pre-processing : "Graph Partitioning"

* Reference : http://ceur-ws.org/Vol-1313/paper_13.pdf

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; With some research we came across a way to reduce query times by storing pre computed paths. However, storing an all pairs path data structure is out of the quesition as the size comeplxity of this algorithm would require an enormous ammount of space. We can however choose certain nodes and decide which paths amongst these nodes should be computed and graph partitioning introduces a way to accomplish this task.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Graph partitioning relies on the idea that if a graph were to be divided into sections where these sections are connected by the fewest number of edges, then these sections must be traversed by the edges that connect them. Then the nodes that connect these edges will almost always be accessed in the query and pre-processing the paths to these nodes along the border of each section could prove to useful while being manageable in size. Then our query times would be reduced and this is trivially true because it would be akin to finding a path along two points while alredy having some of the work done and stored beforehand. To address the issue of dynamic traffic, these path would have to be pre-processed every time there are major changes to the paths that exist amongst the designated border nodes.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; The time complexity of A* Searching algorithm is entirely dependent on the heuristic function it uses to estimate the distance between two vertices. In the worst case, in the unbound tree data structure, the complexity of finding the shortest path can be $O(b^d)$ where $b$ is a branching factor. The best heuristic is one with b = 1. This heuristic function estimates the distance between two nodes by selecting the landmark nodes in the graph and using the triangle inequality. The dillemma between 'speed' and 'memory' is always underlining in the problem of finding the shortest path using ${A*}$ searching Algorithm. In order to significantly reduce the running time of the ${A*}$ searching algorithm, the distance between any two vertices can be estimated using the difference of the distances to their closest landmark. By using the triangle inequality, we can obtain the lower bound of the distance between these vertices and use this lower bound as a heuristic value to improve the time complexity of our implementation. The only issue with this method is the memory allocation of pre-processed distances between a node and any landmark, which may prove to be a problem in terms of size when it comes to larger graphs. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; But more importantly in a dynamic setting and also in our case of finding the fastest path rather than the shortest, this heuristic will prove to be the most accurate, as we will now be able to implement weights with a time metric as we need.

### Partitioning the Graph

* To make $n^2$ number of landmarks, they would be pre-assigned as the centers of $n^2$ partition.
* The partitioning phase determines which nodes will be landmarks by evenly placing squares around the map and choosing the node closest to the center of those squares
* We then compute and store the distances from each node to each landmark.

```python
import math

class partitions:
    # _max is a tuple of the maximum xy-coordinate
    # _min is a tuple of the minimum xy-coordinate
    # n_sq is a squared value of i where i = 1,2,3,4,...
    def __init__(self, _max, _min, n_sq, nodes):
        x_partition = []
        y_partition = []
        # make the n^2 partititions
        x_len = _max[0]-_min[0]
        y_len = _max[1]-_min[1]
        n = math.sqrt(n_sq)
        for i in range(int(n)+1):
            x_partition.append(_min[0] + i*(x_len/n))
            y_partition.append(_min[1] + i*(y_len/n))
        
        landmark = [] # store node key that is landmark
        center_coord=[] # the coord of center of each partitions
        # the node that closest to the center of each partitions is the landmarks
        for i in range(int(n)):
            for j in range(int(n)):
                center_coord.append(((x_partition[i] + x_partition[i+1])/2,(y_partition[j] + y_partition[j+1])/2))
        for center in center_coord:
            dist='inf'
            x=center[0]
            y=center[1]
            for node in nodes:
                lm_x=node.x_coord
                lm_y=node.y_coord
                new_dist = math.sqrt((x-lm_x)**2 + (y-lm_y)**2)
                if new_dist < dist:
                    dist = new_dist
                    landmark_node=node
            landmark.append(landmark_node.key)
        landmark=sorted(landmark)
        
        landmark_final = []
        for k in range(len(landmark)):
            add = (k, landmark[k])
            landmark_final.append(add)  # (index, landmark key)
        
        self.x_partition = x_partition
        self.y_partition = y_partition
        self.landmark = landmark_final
        self.nodes=nodes
    
    # to learn which landmark a node is closest to 
    def which_landmark(self, node):
        landmark_index = -1
        dist = 'inf'
        x = node.x_coord
        y = node.y_coord
        for i in range(len(self.landmark)):
            lm_x = nodes[self.landmark[i][1]].x_coord
            lm_y = nodes[self.landmark[i][1]].y_coord
            
            # eucledean distance between a landmark and a node
            new_dist = math.sqrt((x-lm_x)**2 + (y-lm_y)**2)
            #print new_dist, dist          # for debugging
            if new_dist < dist:
                dist = new_dist
                landmark_index = i
            
        return landmark_index
```

### Implementation of Landmark to Node Distances

* Here we tried doing a brute force calculation of all distance from each node to each landmark.
    * This took anywhere from 25 to 28 seconds.
* We then decided that we could use the concept of memoization to reduce the run time of this algorithm.
    * Based on the graph theory that any sub-path within a path is also a fastest path.
    * We were able to reduce run times to under 0.5 seconds.

```python
import numpy as np
from heapq import heappush, heappop

def dijkstra_d(nodes,start,landmark): #end): # input: start = node key, landmark = (index, landmark key)
    distances = [float('inf') for i in range(len(nodes))] #initialize distances
    distances[start]=0
    que=[(distances[start],start)]
    while que:
        state=heappop(que)[1]
        if state == landmark[1]:   # landmark = (index, landmark key)
            return distances[state]
        if nodes[state].landmark_distances[landmark[0]][0] != -1:                # dynamic programming
            return distances[state] + nodes[state].landmark_distances[landmark[0]][1]    # add distance and return if distance was already computed
        for neigh in nodes[state].neighbors_and_weights:
            dist_from_state=distances[state]+neigh[1] # neigh[1] = weight of neigh
            if dist_from_state<distances[neigh[0]]:
                distances[neigh[0]]=dist_from_state                
                if distances[neigh[0]]<float('inf'):
                    heappush(que,(distances[neigh[0]],neigh[0]))

# a distance from the closest landmark to the node
## for each landmark, precompute the distance from and to the node
def distance_landmark_node(xy_landmark, Nodes): # input landmark_node_key
    #d = [[] for i in range(len(Nodes))]     - commented this out for now
    for node in Nodes:
        node.landmark_distances = [(-1, -1) for x in range(len(xy_landmark))] # initialize the landmark_distances in node
    for node in Nodes:
        i = 0      # landmark index
        d = 'inf'  # default infinity
        for landmark in xy_landmark:            # here landmark is the index of nodes
            d_temp = dijkstra_d(Nodes, node.key, landmark)
            node.landmark_distances[i] = (landmark[1], d_temp)   # add (landmark key, distance to landmark)
            if d_temp < d:
                d = d_temp
                node.landmark_closest = i
            i += 1
                    
# lower bound #1: d(v,w) >= d(v,A) + d(A,w) by Triangular Inequality where A is the chosen landmark
# Hence, d(v,w) >= d(A,w) - d(A,v)
# lower bound #2: d(v,w) >= d(A,w) + d(v,A)
# hence, d(v,w) >= d(v,A) - d(w,A)
# ==> lower bound = max{lower_bound_1, lower_bound_2}
def landmark_heuristic(current_node, terminal_node): #, dist_landmark_to_node, closest_landmark_list):
    current = current_node
    goal = terminal_node
    #d = (np.array(dist_landmark_to_node).transpose())    # d[landmark][node]  (input is d[node][landmark])
    
    ### we need to implement the best_landmark function to find the best landmark to use it for heuristic
    ## cncli : current_node_closest_landmark_index (i.e. the index of the closest landmark of the current node)
    ## tncli : terminal_node_closest_landmark_index (i.e. the index of the closest landmark of the terminal node)
    ## cncli = closest_landmark_list[current_node.key]
    #tncli = closest_landmark_list[terminal_node.key]
    
    #print "tncl, tncl_i, goal.key, current.key:", tncl, tncl_i, goal.key, current.key
    #lower_bound1 = abs(d[tncli][goal.key] - d[tncli][current.key])
    
    landmark_key = goal.landmark_closest
    h_score = abs(current.landmark_distances[landmark_key][1] - goal.landmark_distances[landmark_key][1])
    
    return h_score
    #return lower_bound1
```

### Reduction of Optimal Landmark Selection Problem to Maximum Coverage Problem

* Reference : http://www.fabianfuchs.com/fabianfuchs_ALT.pdf   <br>

$\textbf{Idea}:$ <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; We wanted to investigate how the number of landmarks would affect the time complexity of this searching algorithm and give a brief answer to this question: $\text{"What is the optimal number of landmarks for the best performance in terms of time complexity?"}$ <br>


$\textbf{Definition}:$ (MinALT) <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Given a graph $G = (V, E)$ and some integer k, the MinALT problem is to select $k$ landmarks s.t the resulting overall ALT searching space is minimal <br>
<br>
$\textbf{Definition}$: (Maximum Coverage Problem)   <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Given a set of elements $\{e_1, e_2, ..., e_r\}$ and a collection $S = \{S_1, S_2, ..., S_n\}$ of sets where $S_i \subset \{e_1, e_2, ..., e_r\}$ and some integer $k$. The Maximum Cover Problem is to find a subset $S' \subset S$ with $|S'| <= K$ so that a maximum number of elements $e_i$ are covered. 
<br>

$\textbf{Reduction}$: (MinALT to Max. Cover) <br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; For each vertex in the graph and each s-t-$query$, compute which vertices the vertex $v$ would exclude from the Dijkstra search if it would be a landmark. One can approach this problem by computing the search space of Dijkstra Algorithm and creating $|V|^3$x $|V|$ matrix with a special column for vertices in the Dijkstra's searching range and a row for each vertex $v$. For every vertex $v$ in the search space of Dijkstra, we compute for every landmark $l_i$ whether our algorithm would visit vertex $v$ while selecting $l_i$ as our landmarks. <br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Fill up the matrix as follows:- <br>
$\begin{equation}
  entry_{iv}=\begin{cases}
    1, & \text{if not visited}.\\
    0, & \text{otherwise}.
  \end{cases}
\end{equation}
$ where $entry_{iv}$ is the entry for the row $l_i$ in the $v$ column
<br>

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Selecting $k$ rows to cover the maximum number of columns gives a solution to the MinALT problem. Looking at the reduction process and the generated matrix, one can observe that the reduction from MinALT to Max Cover problem can be done in a polynomial complexity, $O(|V|^4)$.
